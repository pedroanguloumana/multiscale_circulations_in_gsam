# Composite Circulations

```python
phase = 4
region = REGION
fn = f'{ProjectConfigs().project_root_dir}/data/moisture_space_composite.circulation.phase{phase:.0f}.{region}.nc'
```

```python
def plot_composite_circulation(ax, phase, region):
    fn = f'{ProjectConfigs().project_root_dir}/data/moisture_space_composite.circulation.phase{phase:.0f}.{region}.nc'
    assert(os.path.exists(fn))
    da = xr.open_dataset(fn)
    x = (da.rho*da.W).mean('observations')#.plot.contour(ax=ax, levels=np.linspace(-150, 0, 25), colors='k')
    return(x)
```

```python
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(10,5))
for i, ax in enumerate(axs.ravel()):
    x = plot_composite_circulation(ax, i+1, REGION)
    break
    ax.set_ylim(0, 17e3)
fig.tight_layout(pad=1)
```
