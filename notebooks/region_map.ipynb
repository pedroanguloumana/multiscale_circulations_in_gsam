```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.patches import Rectangle
from src.configs import ProjectConfigs
```

```python
# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Define a Cartopy projection
ax = plt.axes(projection=ccrs.PlateCarree())

# Set the extent of the map (longitude, latitude)
ax.set_extent([110, 180, -10, 30], crs=ccrs.PlateCarree())

# Add natural features like coastlines and borders
ax.coastlines(resolution='110m')
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Define the rectangle's coordinates: (lower-left corner x, lower-left corner y, width, height)
rect = Rectangle((130, 0), 40, 20, linewidth=2, edgecolor='red', facecolor='none', transform=ccrs.PlateCarree())

# Add the rectangle to the plot
ax.add_patch(rect)

# Add gridlines
ax.gridlines(draw_labels=True, linestyle="--")


plt.savefig(f'{ProjectConfigs().project_root_dir}/figures/tropical_northwest_pacific.pdf', format='pdf')
```
