```python
%load_ext autoreload
%autoreload 2
import xarray as xr
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from src.plotting import *
from src.utils import *
```

```python
gsam_pcs = load_gsam_pcs_on_era5_levels()
era5_pcs = load_era5_pcs().sel(time=gsam_pcs.time)
```

```python
gsam_phase = compute_phase_number(compute_theta(gsam_pcs))
era5_phase = compute_phase_number(compute_theta(era5_pcs))
```

```python
gsam_phase = gsam_phase.where(compute_r(gsam_pcs)>=0.5, 0)
era5_phase = era5_phase.where(compute_r(era5_pcs)>=0.5, 0)
```

# Transition Frequencies

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,5))
ax=axs[0]
initial_phase = gsam_phase.data[0:-3, :, :]
final_phase = gsam_phase.data[2:-1, :, :]
trans_matrix = transition_matrix(initial_phase, final_phase)
sns.heatmap(trans_matrix, ax=ax, annot=True, fmt=".2f", cmap="plasma", cbar=True, norm=colors.Normalize(vmin=0, vmax=.40))
ax.set_title('gSAM')
ax.set_title('a', loc='left')
ax.set_ylabel(r'Phase -3hr')
ax.set_xlabel(r'Phase +3hr')

ax = axs[1]
initial_phase = era5_phase.data[0:-3, :, :]
final_phase = era5_phase.data[2:-1, :, :]
trans_matrix = transition_matrix(initial_phase, final_phase)
sns.heatmap(trans_matrix, ax=ax, annot=True, fmt=".2f", cmap="plasma", cbar=True, norm=colors.Normalize(vmin=0, vmax=.40))
ax.set_title('ERA5')
ax.set_title('b', loc='left')
ax.set_ylabel(r'Phase -3hr')
ax.set_xlabel(r'Phase +3hr')
fig.tight_layout(pad=1)
save_figure(fig, 'transition_frequencies.pdf')
```

# GMS/VADV

```python
gsam_w = load_gsam_coarse_50pix('W').W
gsam_q = load_gsam_coarse_50pix('QV').QV *1e-3 # convert to kg/kg
gsam_pcs = load_gsam_pcs_on_era5_levels()
gsam_2d = load_gsam_coarse_50pix('2D')
gsam_rho = load_gsam_ref_data().rho
gsam_vadv = ((gsam_w*gsam_rho)*gsam_q).integrate('z')
gsam_phase = compute_phase_number(compute_theta(gsam_pcs))
gsam_phase[compute_r(gsam_pcs)<0.5] = 0
```
