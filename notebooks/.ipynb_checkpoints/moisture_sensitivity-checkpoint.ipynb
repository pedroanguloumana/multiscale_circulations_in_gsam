```python
%load_ext autoreload
%autoreload 2
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from xhistogram.xarray import histogram
from src.plotting import *
from src.utils import *
```

```python
gsam_2d = load_gsam_coarse_50pix('2D')

gsam_pr = gsam_2d.Prec
gsam_csf = (gsam_2d.PW/gsam_2d.PWS).rename('CSF')
```

```python
h = histogram(gsam_pr, gsam_csf, bins=[np.logspace(-1, 2.1, 100), np.linspace(0.3, 1, 100)])
h = h.astype(float)
```

```python
fig, ax = plt.subplots()
cmap = plt.cm.plasma.copy()
norm = colors.Normalize(1, 10)
cmap.set_under('white')
h.plot(ax=ax, cmap=cmap, norm=norm, cbar_kwargs={'orientation': 'horizontal'})
ax.set_title('Joint Histogram, Precip vs CSF')
ax2 = ax.twinx()
h.mean('Prec_bin').plot(ax=ax2, color='black', lw=2)
ax2.set_ylim(0, 20)
ax2.set_ylabel('Number of Obs')
save_figure(fig, 'moisture_sensitivity.pdf')
```
