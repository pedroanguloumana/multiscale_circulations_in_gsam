# Gridscale Evolution
In this notebook we investigate the temporal evolution of grid-scale circulations.

```python
%load_ext autoreload
%autoreload 2
import xarray as xr
import numpy as np
from scipy.stats import binned_statistic_2d
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from xhistogram.xarray import histogram
from src.project_configs import *
from src.utils import *
from src.plotting import *
```

```python
era5_pcs = load_era5_pcs()
gsam_pcs = load_gsam_pcs_on_era5_levels()
assert((era5_pcs.time==gsam_pcs.time).all())
```

# Quiver Plots

```python
pc1_bins = np.arange(-10, 10, 0.5)
pc2_bins = np.arange(-10, 10, 0.5)
midpoint_pc1_bins = array_midpoints(pc1_bins)
midpoint_pc2_bins = array_midpoints(pc2_bins)
xmesh, ymesh = np.meshgrid(midpoint_pc1_bins, midpoint_pc2_bins)
normed = True
arrow_space = 2
obs_cutoff=5
```

```python
fig, axs = plt.subplots(ncols=3, nrows=2, figsize=(15, 7))
for axi, ax in enumerate(axs.ravel()):
    i, j = np.unravel_index(axi, axs.shape)
    pcs = [gsam_pcs, era5_pcs][i]
    dataset = ['gSAM', 'ERA5'][i]
    method = ['forward', 'backward', 'centered'][j]
    pivot = ['tail', 'tip', 'mid'][j]
    delta_pcs = compute_mean_pc_tendency(pcs, method, normed)
    hist = binned_statistic_2d(
                pcs.sel(mode=2).data.ravel(),
                pcs.sel(mode=1).data.ravel(),
                None,
                statistic='count', 
                bins=[pc2_bins, pc1_bins]
            ).statistic/delta_pcs.size
    
    pcs = pcs.sel(time=delta_pcs.time)
    bin_mean_data = dict()
    for mode in [1,2]:
        bin_mean_delta_pc = binned_statistic_2d(
            pcs.sel(mode=2).data.ravel(),
            pcs.sel(mode=1).data.ravel(),
            delta_pcs.sel(mode=mode).data.ravel(),
            statistic=lambda x: np.nanmean(x) if x.size>obs_cutoff else np.nan, 
            bins=[pc2_bins, pc1_bins]
        ).statistic
        bin_mean_data[mode] = bin_mean_delta_pc
       
    c = ax.pcolormesh(xmesh, ymesh, 100*hist, norm=colors.LogNorm(vmin=1e-4, vmax=1e0), cmap=plt.cm.plasma.copy())

    samp = (slice(None, None, arrow_space), slice(None, None, arrow_space-1))
    ax.quiver(xmesh[samp], ymesh[samp], bin_mean_data[1][samp], bin_mean_data[2][samp], angles='xy', pivot=pivot, scale=8, width=0.007, headwidth=3, lw=0.009, edgecolor=None, facecolor='grey')
    ax.set_title(chr(ord('a') + axi), loc='left')
    ax.set_title(f'{dataset} {method}')
    ax.set_xlabel(rf'Normed PC_convective')
    ax.set_ylabel(rf'Normed PC_stratiform')
    ax.grid()
    ax.set_xlim(-5.5, 5.5)
    ax.set_ylim(-5.5, 5.5)
    fig.colorbar(c)

plt.tight_layout(pad=1)
save_figure(fig, 'gsam_era5_pc_evolution.pdf')
```

# Heatmaps

```python
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(15, 7))
bins = np.linspace(-np.pi, np.pi, 100)
xmesh, ymesh = np.meshgrid(array_midpoints(bins), array_midpoints(bins))
# plot gsam heatmap
ax = axs[0]
era5_pcs = load_era5_pcs()
gsam_pcs = load_gsam_pcs()
assert((era5_pcs.time==gsam_pcs.time).all())
gsam_r = compute_r(gsam_pcs)
gsam_theta = compute_theta(gsam_pcs)
gsam_theta_initial = gsam_theta.isel(time=slice(0, -3)).assign_coords(time=gsam_theta.time[1:-2])  # 3 hours ago
gsam_theta_final = gsam_theta.isel(time=slice(2, -1)).assign_coords(time=gsam_theta.time[1:-2])    # 3 hours from now
h = histogram(gsam_theta_initial, gsam_theta_final, bins=[bins, bins])
ax.pcolormesh(xmesh, ymesh, h, cmap=plt.cm.plasma.copy())

```
