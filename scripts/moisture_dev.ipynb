```python
import xarray as xr
import numpy as np
import os
from src.configs import Configs
from glob import glob

import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(15, 5))

for i, ax in enumerate(axs.ravel()):
    circ = xr.open_dataarray(f'../data-output/gsam.composite_circulation.phase{i+1:.0f}.nc').coarsen({'column': 5}).mean()
    circ.plot.contourf(ax=ax, levels=np.linspace(-150, 0, 20), cmap=plt.cm.jet)
fig.tight_layout(pad=1.0)
```
