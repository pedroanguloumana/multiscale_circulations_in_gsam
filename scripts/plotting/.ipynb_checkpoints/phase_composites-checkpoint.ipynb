```python
import xarray as xr
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from src.configs import ProjectConfigs
```

```python
REGION = 'northwest_tropical_pacific'
GRIDSIZE = 50
```

```python
def get_phase_composite_filename(var, phase, region):
    return f"{ProjectConfigs().project_root_dir}/data/moisture_space_composite.{var}.phase{phase}.{region}.nc"
```

# Let's look at $\psi$ and $w$ plots

```python
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20, 10))
for i in range(1,9):
    ax = axs[np.unravel_index(i-1, axs.shape)]

   
    # w plots
    w = xr.open_dataset(get_phase_composite_filename('W', i, REGION)).W*100
    w = w.coarsen({'column': 20}).mean()
    levels = np.arange(-5, 15, 0.5)
    w.plot.contourf(ax=ax,
                    levels=levels,
                    cmap=plt.cm.bwr.copy(),
                    norm=colors.TwoSlopeNorm(vcenter=0, vmin=levels.min(), vmax=levels.max()),
                    cbar_kwargs={'label': 'W [cm/s]'})
     
    # circulation plots
    circ = xr.open_dataset(get_phase_composite_filename('circulation', i, REGION)).W
    circ = circ.assign_coords({'column': np.linspace(0, 1, circ.column.size)})
    circ = circ.coarsen({'column': 20}).mean()
    levels = np.linspace(-150, 0, 20)
    circ.plot.contour(ax=ax, levels=levels, colors='black')

    ax.set_ylim(0, 17e3)
    ax.set_title(f'Phase {i}')
    ax.set_ylabel('Altitude [m]')
    ax.set_xlabel('Column Rank')
    
fig.tight_layout(pad=2)
fn_to_save = f'{ProjectConfigs().project_root_dir}/figures/phase_composites_psi_w.pdf'
plt.savefig(fn_to_save, format='pdf')
```

# Now $\psi$ and MSE

```python
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20, 10))
for i in range(1,9):
    ax = axs[np.unravel_index(i-1, axs.shape)]

   
    # MSE plots
    qv = xr.open_dataset(get_phase_composite_filename('QV', i, REGION)).QV * 1e-3 # convert to kg/kg
    T = xr.open_dataset(get_phase_composite_filename('TABS', i, REGION)).TABS
    Lqv_plus_cpT = (2.5e6*qv) + (1004*T)
    MSE = Lqv_plus_cpT + (9.81*Lqv_plus_cpT.z)
    MSE = MSE / 1000 # convert to kJ
    MSE = MSE.coarsen({'column': 20}).mean()
    levels = np.arange(320, 350, 0.5)
    MSE.plot.contourf(ax=ax,
                      levels=levels,
                      cmap=plt.cm.plasma.copy(),
                      norm=colors.Normalize(vmin=levels.min(), vmax=levels.max()), 
                      cbar_kwargs={'label': 'MSE [kJ/kg]'})

    # circulation plots
    circ = xr.open_dataset(get_phase_composite_filename('circulation', i, REGION)).W
    circ = circ.assign_coords({'column': np.linspace(0, 1, circ.column.size)})
    circ = circ.coarsen({'column': 20}).mean()
    levels = np.linspace(-150, 0, 20)
    circ.plot.contour(ax=ax, levels=levels, colors='black')

    ax.set_ylim(0, 17e3)
    ax.set_title(f'Phase {i}')
    ax.set_ylabel('Altitude [m]')
    ax.set_xlabel('Column Rank')
    
fig.tight_layout(pad=2)
fn_to_save = f'{ProjectConfigs().project_root_dir}/figures/phase_composites_psi_mse.pdf'
plt.savefig(fn_to_save, format='pdf')
```

# Now $\psi$ and Clouds

```python
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20, 10))
for i in range(1,9):
    ax = axs[np.unravel_index(i-1, axs.shape)]

   
    # QCpI plots
    qc = xr.open_dataset(get_phase_composite_filename('QC', i, REGION)).QC * 1e-3 # convert to kg/kg
    qi = xr.open_dataset(get_phase_composite_filename('QI', i, REGION)).QI * 1e-3 # convert to kg/kg
    qcpi = qc + qi
    qcpi = qcpi.coarsen({'column': 20}).mean()
    levels = np.linspace(1e-5, 2e-4, 25)
    cmap = plt.cm.plasma.copy()
    cmap.set_under('white')
    c = qcpi.plot.contourf(ax=ax,
                          levels=levels,
                          cmap=cmap,
                          norm=colors.Normalize(vmin=levels.min(), vmax=levels.max()),
                          cbar_kwargs={'label': 'QC+QI [kg/kg]'})


    # circulation plots
    circ = xr.open_dataset(get_phase_composite_filename('circulation', i, REGION)).W
    circ = circ.assign_coords({'column': np.linspace(0, 1, circ.column.size)})
    circ = circ.coarsen({'column': 20}).mean()
    levels = np.linspace(-150, 0, 20)
    circ.plot.contour(ax=ax, levels=levels, colors='black')

    ax.set_ylim(0, 17e3)
    ax.set_title(f'Phase {i}')
    ax.set_ylabel('Altitude [m]')
    ax.set_xlabel('Column Rank')
    
fig.tight_layout(pad=2)
fn_to_save = f'{ProjectConfigs().project_root_dir}/figures/phase_composites_psi_qcpi.pdf'
plt.savefig(fn_to_save, format='pdf')
```
